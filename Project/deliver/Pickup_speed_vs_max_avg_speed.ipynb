```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = pd.read_pickle('../data/wss')
```

```python
df = df[['itapudid', 'avgspeed', 'max1stdetectwssc','maxspeed','tripdist','eventtime']]
```

```python
df = df[df['itapudid'] == '170540048001DC92C8827']
```

```python
plt.figure(figsize=(20, 5));
df = df.sort_values(['eventtime']).dropna(axis=0, how='any').reset_index(drop = True)
df = df[df['eventtime']!='1970-01-01'].reset_index(drop = True)
df = df.set_index(pd.DatetimeIndex(df['eventtime']))
plt.subplot(1,3,1)
plt.plot(df.index.values, df['max1stdetectwssc'], label='max1stdetectwssc')
plt.legend()
plt.subplot(1,3,2)
plt.plot(df.index.values, df['avgspeed'], label='avgspeed')
plt.legend()
plt.subplot(1,3,3)
plt.plot(df.index.values, df['maxspeed'], label='maxspeed')
plt.legend()
```
