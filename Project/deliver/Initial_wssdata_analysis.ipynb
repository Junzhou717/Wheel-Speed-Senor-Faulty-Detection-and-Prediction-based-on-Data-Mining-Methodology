```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_pickle('../data/wss_manufacturer_merge')
```

```python
def if_na(maxc, maxd, maxe, maxf):
    nac = 0;
    if maxc == 0: nac += 1
    if maxd == 0: nac += 1
    if maxe == 0: nac += 1
    if maxf == 0: nac += 1
    if nac == 2:
        return 0
    elif nac == 1:
        return 1
    else:
        return 2
```

```python
df = df.reset_index(drop=True)
```

```python
df = df.fillna(0)
```

```python
df = pd.DataFrame(df, columns=['itapudid', 'max1stdetectwssc', 'max1stdetectwssd', 'max1stdetectwsse', 'max1stdetectwssf', 'eventtime', 'manufacturer', 'na_value'])
```

```python
import pandasql as pdsql

pysql = lambda q: pdsql.sqldf(q, globals())
str1 = """select 
            itapudid, manufacturer, sum(na_value), count(na_value)
        from 
            df
        group by
            itapudid
            """

df1 = pysql(str1)
```

```python
df1 = pd.DataFrame(df1, columns=['itapudid', 'manufacturer', 'sum(na_value)', 'count(na_value)', 'na_value_ratio'])
```

```python
for i in range(171):
    df1.set_value(i,'na_value_ratio',(df1.get_value(i,'sum(na_value)'))/(df1.get_value(i,'count(na_value)')*2))  
```

```python
df1 = df1[['itapudid', 'na_value_ratio', 'manufacturer']]
```

```python
df2 = pd.read_pickle('../data/wss_n_c1')
df2 = df2[df2.eventtime != '1970-01-01']
grouped = df2.groupby('itapudid')
df2 = grouped.describe().reset_index()
```

```python
df2_1 = df2.itapudid
```

```python
df2_2 = df2.speed
```

```python
df2_c = pd.concat([df2_1, df2_2], axis=1)
```

```python
df2_c = df2_c[['itapudid', 'mean', 'max', 'std']]
```

```python
df_mer = pd.merge(df1, df2_c, how='left', on='itapudid')
```

```python
df_mer = df_mer.fillna(0)
```

```python
df_mer = df_mer[df_mer['max'] != 0]
```

```python
df_mer.to_pickle('../data/wss_manufacturer_classify')
```
