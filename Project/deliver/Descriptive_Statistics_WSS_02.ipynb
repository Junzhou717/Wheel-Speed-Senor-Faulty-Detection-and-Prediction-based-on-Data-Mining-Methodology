### Findings
From the 4 output figures above, they show that there is high frequency of pick-up speed of 0. It is unnormal and we will figure it out in later analysis.

```python
grouped = df.groupby('itapudid') # Group the dataset by itapudid
grouped.describe()
```

```python
df_mean = df.groupby('itapudid').describe().reset_index()
```

```python
df_mean = df_mean.speed
df_mean = df_mean.rename(columns = {'mean': 'speed'})
```

```python
plt.hist(df_mean.speed, bins=np.arange(df_mean.speed.min(), df_mean.speed.max()+1))
plt.title('WSS Mean Speed Distribution')
plt.ylabel('Frequency')
plt.xlabel('Speed')
plt.show()
```

The figure above shows that the average pickup speed follow two Gaussian distributions. In later analysis, we found the reason is that they are from two different manufacturers.
