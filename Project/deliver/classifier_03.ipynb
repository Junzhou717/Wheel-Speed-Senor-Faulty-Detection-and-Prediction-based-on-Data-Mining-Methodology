### Description
This document gives the classifier which used to determine whether the values are collected correctly. If it is, return 1. Otherwise return 0. Normally, 2 or 4 speed values should be collected. There are 3 patterns: sensor C and D, sensor E and F, sensor C, D, E and F. The values in each patterns should be more than 0 and less than 120.

```python
import pandas as pd
df = pd.read_pickle('../data/wss')
```

```python
df
```

```python
def outlier(max1stdetectwssc, max1stdetectwssd, max1stdetectwsse, max1stdetectwssf):
    if ((max1stdetectwssc > 0) & (max1stdetectwssd > 0) & (max1stdetectwsse > 0) & (max1stdetectwssf > 0)):
        if ((max1stdetectwssc <= 120) & (max1stdetectwssd <= 120) & (max1stdetectwsse <=120) & (max1stdetectwssf <= 120)):
            return 1
        else:
            return 0
    elif ((max1stdetectwssc > 0) & (max1stdetectwssd > 0)):
        if (max1stdetectwssc <= 120) & (max1stdetectwssd <= 120):
            return 1
        else:
            return 0
    elif ((max1stdetectwsse > 0) & (max1stdetectwssf > 0)):
        if ((max1stdetectwsse <= 120) & (max1stdetectwssf <= 120)):
            return 1
        else:
            return 0
    else:
        return 0 
```

```python
results =df[['itapudid', 'max1stdetectwssc', 'max1stdetectwssd', 'max1stdetectwsse', 'max1stdetectwssf', 'eventtime']]
```

```python
results = results.fillna(0)
```

```python
results1 = pd.DataFrame(results, columns=['itapudid', 'max1stdetectwssc', 'max1stdetectwssd', 'max1stdetectwsse', 'max1stdetectwssf', 'eventtime', 'correct'])
```

```python
for i in range(100000):
    results1.set_value(i,'correct',outlier(results1.get_value(i,'max1stdetectwssc'),results1.get_value(i,'max1stdetectwssd'),results1.get_value(i,'max1stdetectwsse'),results1.get_value(i,'max1stdetectwssf')))    
```
