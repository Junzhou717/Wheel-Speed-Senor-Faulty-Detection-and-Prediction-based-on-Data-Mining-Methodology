### Description
This document explains how to merge the excel file with dataframe using pandas. After doing this, each poweroffovent will have a new feature named manufacturer. The poweroffevents which haven't manufacturers provided by the company will be removed.

Additionally, a new dataset with features named itapudid, na_value_ratio, std, mean, max, manufacturer is created at last. The na_value_ratio describes the NAN value ratio of each itapudid.

```python
import pandas as pd
import math
df = pd.read_excel('../data/manufacturer_labels1.xlsx')
```

```python
df1 = df[['itapudid', 'manufacturer']]
```

```python
df_wss_na = pd.read_pickle('../data/na_ratio_wss')
```

```python
#merge the excel data with the sql dataset
df_mer = pd.merge(df_wss_na, df1, how='left', on='itapudid')
```

```python
df_mer = df_mer[(df_mer['manufacturer'] == 'Knorr Bremse')|(df_mer['manufacturer'] == 'Haldex')|(df_mer['manufacturer'] == 'Wabco')]
```

```python
df_mer
```

```python
df_mer.to_pickle('../data/wss_manufacturer_merge')
```

```python
df2 = pd.read_pickle('../data/wss_n_c1')
grouped = df2.groupby('itapudid')
df2 = grouped.describe().reset_index()
```

```python
df2_1 = df2.itapudid
df2_2 = df2.speed
df2_c = pd.concat([df2_1, df2_2], axis=1)
df2_c = df2_c[['itapudid', 'mean', 'max', 'std']]
df_mer = pd.merge(df_mer, df2_c, how='left', on='itapudid')
df_mer = df_mer.fillna(0)
df_mer = df_mer[df_mer['max'] != 0]
df_mer.to_pickle('../data/wss_manufacturer_classify')
```
